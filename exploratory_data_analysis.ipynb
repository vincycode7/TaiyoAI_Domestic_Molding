```python
from utils.util_functions import target_correlation_chart
```

```python
corr_target_timeline = ["target_shift_t+1", "target_shift_t+2", "target_shift_t+3"]
target_corr_with_all_feat = target_correlation_chart(processor.transformed_X_data, processor.transformed_Y_data[corr_target_timeline])[:-len(corr_target_timeline)]
target_corr_with_all_feat
```

##### Step4 - Get filtered features with very positive or very negative correlation to the target

```python
target_with_corr_filtered = target_corr_with_all_feat[(target_corr_with_all_feat > 0.4) | (target_corr_with_all_feat < -0.4)].dropna()[:-len(target_corr_with_all_feat.columns)]
target_with_corr_filtered
```

```python
target_with_corr_filtered_excluded = target_corr_with_all_feat[(target_corr_with_all_feat < 0.4) & (target_corr_with_all_feat > -0.4)].dropna()[:-len(target_corr_with_all_feat.columns)]
target_with_corr_filtered_excluded
```

```python
#save all_columns_target_with_corr_filtered
with open("configs/all_columns_for_filtered_corr.yml", "w") as f:
    f.write(f"target_with_corr_filtered_all_columns : {list(target_with_corr_filtered.index)}")
```

```python
#save all_columns_target_with_corr_filtered
with open("configs/all_columns_for_corr_filtered_excluded.yml", "w") as f:
    f.write(f"target_with_corr_filtered_excluded_all_columns : {list(target_with_corr_filtered_excluded.index)}")
```

```python
### Train Data Pipeline
data_postprocessor_config["all_columns"] = all_columns_for_filtered_corr
data_postprocessor_config["split_data"] = False
data_postprocessor_config["save_split_data"] = False
data_postprocessor_config["data_pipelined_path"] = "artifacts/data_pipeline/data_pipeline_with_filtered_corr_feat.pkl"
data_postprocessor_config["data_file_path"] = "artifacts/data/processed/train_split.csv"
processor_train_with_filtered_corr_feat = DataPostProcessor(**data_postprocessor_config)
processor_train_with_filtered_corr_feat.post_process_data(**data_postprocessor_config)

# Test Data Pipeline
data_postprocessor_config["all_columns"] = all_columns_for_filtered_corr
data_postprocessor_config["split_data"] = False
data_postprocessor_config["save_split_data"] = False
data_postprocessor_config["data_pipelined_path"] = "artifacts/data_pipeline/data_pipeline_with_filtered_corr_feat.pkl"
data_postprocessor_config["data_file_path"] = "artifacts/data/processed/test_split.csv"
processor_test_with_filtered_corr_feat = DataPostProcessor(**data_postprocessor_config)
processor_test_with_filtered_corr_feat.post_process_data(**data_postprocessor_config)
```

#### Train Univariant Model (Using various supported model backbones)

```python
from model import LazyPredictRegressionBackBone
feature_timeline = "domestic_market_contract_blow_molding_low"
target_timeline = "target_shift_t+1"
regressor_T1 = LazyPredictRegressionBackBone()
regressor_T1.train(processor_train.transformed_X_data_norm[[feature_timeline]],processor_test.transformed_X_data_norm[[feature_timeline]], processor_train.transformed_Y_data[[target_timeline]], processor_test.transformed_Y_data[[target_timeline]])
regressor_T1.get_evaluation()
```

```python
best_lazy_model = regressor_T1.lazyobject.models["TransformedTargetRegressor"]
# Generate predictions using the trained model
y_pred = best_lazy_model.predict(processor_test.transformed_X_data_norm[[feature_timeline]])

# Calculate evaluation metrics
adj_r2 = 1 - (1 - r2_score(processor_test.transformed_Y_data[[target_timeline]], y_pred)) * ((len(processor_test.transformed_Y_data[[target_timeline]]) - 1) / (len(processor_test.transformed_Y_data[[target_timeline]]) - processor_test.transformed_Y_data[[target_timeline]].shape[1] - 1))
r2 = r2_score(processor_test.transformed_Y_data[[target_timeline]], y_pred)
rmse = np.sqrt(mean_squared_error(processor_test.transformed_Y_data[[target_timeline]], y_pred))
directional_accuracy = accuracy_score(processor_test.transformed_Y_data[[target_timeline]].values > processor_test.transformed_X_data[[feature_timeline]].values, y_pred.reshape(-1, 1) > processor_test.transformed_X_data[[feature_timeline]].values)
f1 = f1_score(processor_test.transformed_Y_data[[target_timeline]].values > processor_test.transformed_X_data[[feature_timeline]].values, y_pred.reshape(-1, 1) > processor_test.transformed_X_data[[feature_timeline]].values)

# Print the evaluation metrics
print("Adjusted R-Squared:", adj_r2)
print("R-Squared:", r2)
print("RMSE:", rmse)
print("Directional Accuracy:", directional_accuracy)
print("F1 Score:", f1)
```
