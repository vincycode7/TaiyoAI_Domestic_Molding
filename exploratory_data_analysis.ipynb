```python
# processor_train.transformed_X_data.columns
def run_lazy_predict_uni_multivariant(feature_timeline=None,target_timeline = ["target_shift_t+1"], time_name="_T+1"):
    from model import LazyPredictRegressionBackBone
    feature_timeline = feature_timeline if feature_timeline is not None else ["domestic_market_contract_blow_molding_low", "target_change", "target_growth", "target_ma_2", "target_ma_3", "target_ma_5", "target_ma_7", "target_ma_10", "target_lag_t-1", "target_lag_t-2", "target_lag_t-3", "target_lag_t-4", "target_lag_t-5", "target_lag_t-6"]
    # processor_train.transformed_X_data[feature_timeline]
    regressor_T1 = LazyPredictRegressionBackBone()
    regressor_T1.train(processor_train.transformed_X_data_norm[feature_timeline],processor_test.transformed_X_data_norm[feature_timeline], processor_train.transformed_Y_data[target_timeline], processor_test.transformed_Y_data[target_timeline])
    # regressor_T1.get_evaluation()
    import json
    result = []

    for each_model in regressor_T1.lazyobject.models.keys():
        best_lazy_model = regressor_T1.lazyobject.models[each_model]
        # Generate predictions using the trained model
        y_pred = best_lazy_model.predict(processor_test.transformed_X_data_norm[feature_timeline])

        # Calculate evaluation metrics
        n = len(processor_test.transformed_Y_data[target_timeline])
        p = processor_test.transformed_Y_data[target_timeline].shape[1]
        adj_r2 = 1 - (1 - r2_score(processor_test.transformed_Y_data[target_timeline], y_pred)) * ((n - 1) / (n - p - 1))
        r2 = r2_score(processor_test.transformed_Y_data[target_timeline], y_pred)
        rmse = np.sqrt(mean_squared_error(processor_test.transformed_Y_data[target_timeline], y_pred))
        mape = mean_absolute_percentage_error(processor_test.transformed_Y_data[target_timeline], y_pred)
        directional_accuracy = accuracy_score(processor_test.transformed_Y_data[target_timeline].values > processor_test.transformed_X_data["domestic_market_contract_blow_molding_low"].values, y_pred.reshape(-1, 1) > processor_test.transformed_X_data["domestic_market_contract_blow_molding_low"].values)
        f1 = f1_score(processor_test.transformed_Y_data[target_timeline].values > processor_test.transformed_X_data[["domestic_market_contract_blow_molding_low"]].values, y_pred.reshape(-1, 1) > processor_test.transformed_X_data[["domestic_market_contract_blow_molding_low"]].values)

        # Print the evaluation metrics
        result.append({"model_name":each_model+time_name,"Adjusted R-Squared:": adj_r2, "R-Squared:": r2, "RMSE:": rmse,"Directional Accuracy:": directional_accuracy, "MAPE": mape})

    return regressor_T1, pd.DataFrame(result).sort_values(by=["RMSE:","Directional Accuracy:","MAPE","R-Squared:", "Adjusted R-Squared:"])

def ihs_forecast(sklearn_model, feature_timeline,file_path="artifacts/data/processed/val_data_1.csv"):
    """_summary_

    Args:
        sklearn_model (_type_): Will use the sklearn_model to make the predictions
        file_path (_type_): Will load the file and process it for predictions
    """
    from configs.configs import get_data_postprocessor_config
    from data.data_preprocessing import DataPostProcessor
    from model import LSTMBackBone
    
    data_postprocessor_config = get_data_postprocessor_config()
    with open("configs/all_columns_for_filtered_corr.yml", "r") as f:
        all_columns_for_filtered_corr = yaml.safe_load(f)["target_with_corr_filtered_all_columns"]
    
    #Val
    data_postprocessor_config["split_data"] = False
    data_postprocessor_config["save_split_data"] = False
    data_postprocessor_config["data_pipelined_path"] = "artifacts/data_pipeline/data_pipeline_with_all_feat.pkl"
    
    # load data, run first predict, update dataframe with new prediction, post proecss, run prediction for future - 1 times then return outputs
    
    data_postprocessor_config["data_file_path"] = file_path

    processor_val = DataPostProcessor(**data_postprocessor_config)
    data_postprocessor_config["all_columns"] = all_columns_for_filtered_corr
    processor_val.post_process_data(**data_postprocessor_config)
    
    # First prediction
    y_pred = sklearn_model.predict(processor_val.transformed_X_data_norm[feature_timeline])
    
    # Update dataframe with new prediction
    
    # Post processing
    
    # Run prediction for future - 1
    
    # return results
    
    return pd.DataFrame(y_pred), processor_val
```
